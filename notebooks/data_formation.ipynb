```python
# data to be feed in make_csv

# creating a list of dictionaries which then will be fed to create a DataFrame()
similar_data = []
for i in range(df.shape[0]):
    if len(df['similar'][i]) == 0:
        continue
    else:
        similar_data.append(df['similar'][i][0])
```

```python
def make_csv(dataframe):
    '''
    dataframe : list of dictionaries
    '''
    # metadata.csv file
    data = pd.DataFrame(dataframe)
    data.drop(columns = ['entities']).to_csv(md_s, na_rep = 'nan', columns = columns)   # inplace = False
    
    # entities.csv file
    ents = []
    for i in range(data.shape[0]):
        for j in range(len(data['entities'][i])):
            data['entities'][i][j]['uuid'] = data['uuid'][i]
            ents.append(data['entities'][i][j])
            
    # creating entitie dataframe
    entitie = pd.DataFrame(ents)
    e = entitie.drop(columns = 'highlights')
    e.to_csv(ent_s, na_rep = 'nan', columns = e.columns)     # entities.csv file created
    
    # highlights.csv file
    for i in range(data.shape[0]):
        for j in range(len(data['entities'][i][0]['highlights'])):
            data['entities'][i][0]['highlights'][j]['uuid'] = data['uuid'][i]   # adding new key "uuid" to the highlights
    
    
    hglts = []

    for i in range(data.shape[0]):
        for u in range(len(data['entities'][i][0]['highlights'])):
            hglts.append(data['entities'][i][0]['highlights'][u])
    
    # creating highlights
    pd.DataFrame(hglts, columns = h_cols).to_csv(high_s, na_rep = 'nan', columns = h_cols)
```

```python
make_csv(similar_data)
```
